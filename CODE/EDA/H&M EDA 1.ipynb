```python
def plot_data(data, column):
    """
    Function that takes in dataset and column as input and plots 
    frequency bar chart and pie chart for the same
    """
    counts = dict(data[column].value_counts())
    print(data[column].value_counts())
    print()

    figure = plt.figure(figsize=(11,5))

    # bar chart
    plt.subplot(1,2,1)
    ax = sns.countplot(x = data[column])
    ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)

    # pie chart
    plt.subplot(1,2,2)
    plt.pie(x=counts.values(), labels=counts.keys())

    plt.show()
```

```python
plot_data(articles, 'index_group_name');
```

```python
plot_data(data=articles, column='garment_group_name')
```

```python
transactions['year'] = pd.DatetimeIndex(transactions['t_dat']).year
transactions['month'] = pd.DatetimeIndex(transactions['t_dat']).month
```

```python
transactions
```

```python
# extacting months of the transactions for that particular year
def extract_x_values (data, year):
    year = data[data['year']==year]
    return year['month'].unique()

# extracting counts of the transactions for all the months for that particular year
def extract_y_values (data, year):
    year = data[data['year']==year]
    count = year['month'].groupby(year.month).agg('count')
    return count
```

```python
from matplotlib.pyplot import figure
figure(figsize=(15, 8), dpi=150)

# Define data values
x_2018 = extract_x_values (transactions, 2018)
y_2018 = extract_y_values (transactions, 2018)
x_2019 = extract_x_values (transactions, 2019)
y_2019 = extract_y_values (transactions, 2019)
x_2020 = extract_x_values (transactions, 2020)
y_2020 = extract_y_values (transactions, 2020)

# Plot 2018 transactions
plt.plot(x_2018, y_2018, 'k', label='2018')

# Plot 2019 transactions
plt.plot(x_2019, y_2019, 'r', label='2019')

# Plot 2020 transactions
plt.plot(x_2020, y_2020, 'y', label='2020')

plt.ticklabel_format(style='plain')
plt.legend(loc='best')
plt.title('Month vs. Number of Transactions')
plt.xlabel('Month')
plt.ylabel('Number of Transactions')
plt.show()

#fig.savefig('Monthvs.NumofTransactions.jpg')
```
